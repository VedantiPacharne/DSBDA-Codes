```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Define the file path
file_path = 'C:/Users/HP/Downloads/iris/iris.data'  # Replace with your actual file path

# Define column names
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

# Load dataset
df = pd.read_csv(file_path, header=None, names=columns)

# Display first few rows explicitly
print(df.head())

# Set the plot style
sns.set(style="whitegrid")

# --- Histograms with labeled axes ---
features = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width']
plt.figure(figsize=(12, 10))

for i, feature in enumerate(features):
    plt.subplot(2, 2, i + 1)
    plt.hist(df[feature], bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Histogram of {feature}')
    plt.xlabel(f'{feature} (cm)')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.suptitle('Histograms of Iris Dataset Features', fontsize=16, y=1.03)
plt.show()

# --- Box plots with labeled axes ---
plt.figure(figsize=(12, 8))
sns.boxplot(data=df[features], palette='pastel')
plt.title('Box Plots of Iris Dataset Features')
plt.xlabel('Feature')
plt.ylabel('Measurement (cm)')
plt.show()
```
