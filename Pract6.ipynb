```python
# Part 1: Load the Titanic dataset and explore patterns using Seaborn

# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the built-in Titanic dataset from Seaborn
titanic = sns.load_dataset('titanic')

# Display the first few rows of the dataset
print("Preview of Titanic Dataset:")
display(titanic.head())

# Dataset structure and summary
print("\nDataset Information:")
titanic.info()

# Check for missing values
print("\nMissing Values in Each Column:")
print(titanic.isnull().sum())

# Pattern 1: Count of passengers by sex
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic, x='sex', palette='pastel')
plt.title('Passenger Count by Gender')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()

# Pattern 2: Survival count by class
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic, x='class', hue='survived', palette='muted')
plt.title('Survival Count by Passenger Class')
plt.xlabel('Class')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No (0)', 'Yes (1)'])
plt.show()

# Pattern 3: Survival by gender with clear legend
plt.figure(figsize=(6, 4))
sns.countplot(data=titanic, x='sex', hue='survived', palette='Set2')
plt.title('Survival by Gender')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['No (0)', 'Yes (1)'])  # Custom legend labels
plt.show()

# Calculate the survival rate for each class
survival_rate_by_class = titanic.groupby('class')['survived'].mean()

# Plot survival rate by class
plt.figure(figsize=(6, 4))
sns.barplot(x=survival_rate_by_class.index, y=survival_rate_by_class.values, palette='muted')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Class')
plt.ylabel('Survival Rate')
plt.show()

# Print survival rates for each class
print("\nSurvival Rate by Class:")
print(survival_rate_by_class)



# Part 2: Plot a histogram for the distribution of ticket prices (fare)

# Plot histogram of 'fare' column
plt.figure(figsize=(10, 6))
sns.histplot(titanic['fare'], kde=True, bins=30, color='skyblue')
plt.title('Distribution of Fare Prices on Titanic')
plt.xlabel('Fare')
plt.ylabel('Number of Passengers')
plt.grid(True)
plt.show()


```
