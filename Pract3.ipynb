```python
# Part 1: Statistical analysis of an Employee Dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math

# Load the dataset (change the path accordingly)
employee_data = pd.read_csv(r"C:/Users/HP/Downloads/archive (6)/Employee.csv")  # e.g., 'employee.csv'

# Display the first few rows
print("First 5 rows of the dataset:")
print(employee_data.head())

# Summary statistics
print("\nStatistical Summary:")
print(employee_data.describe())

# Measures of Central Tendency and Dispersion
print("\nMean:\n", employee_data.mean(numeric_only=True))
print("\nMedian:\n", employee_data.median(numeric_only=True))
print("\nMinimum:\n", employee_data.min(numeric_only=True))
print("\nMaximum:\n", employee_data.max(numeric_only=True))
print("\nStandard Deviation:\n", employee_data.std(numeric_only=True))

# Improved Histograms
numeric_cols = employee_data.select_dtypes(include=np.number).columns

plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_cols):
    plt.subplot(math.ceil(len(numeric_cols)/3), 3, i+1)
    sns.histplot(employee_data[col], kde=True, color='skyblue', bins=15)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.tight_layout()
plt.suptitle("Improved Histograms of Numerical Features", fontsize=16, y=1.02)
plt.show()


# Improved Boxplots
plt.figure(figsize=(15, 10))
for i, col in enumerate(numeric_cols):
    plt.subplot(math.ceil(len(numeric_cols)/3), 3, i+1)
    sns.boxplot(x=employee_data[col], color='lightgreen')
    plt.title(f'Boxplot of {col}')
    plt.xlabel(col)
    plt.tight_layout()
plt.suptitle("Improved Boxplots of Numerical Features", fontsize=16, y=1.02)
plt.show()

```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Define column names
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

# Load dataset
df = pd.read_csv('C:/Users/HP/Downloads/iris/iris.data', header=None, names=columns)

# Display first few rows explicitly
print(df.head())

# Filter only 'Iris-setosa' and 'Iris-versicolor'
filtered_df = df[df['species'].isin(['Iris-setosa', 'Iris-versicolor'])]

# Display statistical details
print("\nDescriptive Statistics:\n", filtered_df.describe(percentiles=[.25, .5, .75]))

# Calculate and display correlation matrix
correlation = filtered_df.drop(columns='species').corr()
print("\nCorrelation Matrix:\n", correlation)

# Plot heatmap of correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap of Iris-setosa and Iris-versicolor')
plt.show()
```
