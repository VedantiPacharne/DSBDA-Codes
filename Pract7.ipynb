```python
# Improved Box Plot: Age Distribution by Sex and Survival (Titanic Dataset)

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Titanic dataset from seaborn
titanic = sns.load_dataset('titanic')

print("Preview of Titanic Dataset:")
display(titanic.head())

# Drop rows with missing age values
titanic_clean = titanic.dropna(subset=['age'])







# Set a visually appealing style
sns.set(style="whitegrid")

# Define a custom color palette using color names
custom_palette = {0: "salmon", 1: "mediumseagreen"}  # red-like for did not survive, green-like for survived

# Create the boxplot
plt.figure(figsize=(10, 6))
ax = sns.boxplot(
    data=titanic_clean,
    x='sex',
    y='age',
    hue='survived',
    palette=custom_palette
)

# Customize legend with clearer labels
handles, labels = ax.get_legend_handles_labels()
ax.legend(handles=handles, labels=['Did Not Survive (0)', 'Survived (1)'], title='Survival Status')

# Add plot titles and labels
plt.title('Age Distribution by Sex and Survival on the Titanic', fontsize=14)
plt.xlabel('Sex', fontsize=12)
plt.ylabel('Age', fontsize=12)
plt.tight_layout()

# Show the plot
plt.show()
```
