```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("C:/Users/HP/Downloads/archive (5)/StudentsPerformance.csv")

# Now display 
df.head()

# Listing all column names next cell
print(df.columns.tolist())

# Checking missing values next cell 
df.isnull().sum()

# Fill missing values if necessary
df.fillna(method='ffill', inplace=True)

# Check duplicates
df.duplicated().sum()

# Remove duplicates
df = df.drop_duplicates()

# See all columns
print(df.columns)

# Set the style for plots
sns.set(style="whitegrid")

# 1. Distribution of Scores (Math, Reading, Writing) - Vertical Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['math score', 'reading score', 'writing score']], palette='Set2')
plt.title('Distribution of Scores')
plt.show()





# 3. Test Preparation Course with custom colors
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='test preparation course', palette='Set2')  # Custom color palette
plt.title('Test Preparation Course Distribution')
plt.show()


# See all columns
print(df.columns)


# Correct list of categorical columns
categorical_cols = ['gender', 'race/ethnicity', 'parental level of education', 'lunch', 'test preparation course']

# Apply Label Encoding
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

# Display updated dataframe
df.head()
```
